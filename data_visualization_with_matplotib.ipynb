```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# importing the dataset
kashti = sns.load_dataset("titanic")
```

```python
kashti.info()
```

```python
# Checking the columns with null values
kashti.isnull().sum()
```

```python
# Creating the bar plot
kashti.sex.value_counts().plot(kind="bar")
```

```python
# creating scatter plot by matplotlib
plt.scatter(data=kashti, y="age", x="fare", c="pclass")
```

```python
# Creating scatter plot by seaborn
sns.scatterplot(data=kashti, x="fare", y="age", hue="pclass")
```

```python
# Pair plot by seaborn
sns.pairplot(kashti)
```

```python
# Histogram plot
kashti.hist(grid=False, figsize=(10,9))
```

```python
# Creating the box plot
sns.boxplot(data=kashti, x="sex", y="fare", hue="pclass")
```

```python
#creating the facegrid map
g = sns.FacetGrid(kashti, hue="survived", col="who",margin_titles=True,
                  palette={1:"yellow",0:"green"})
g = g.map(plt.scatter, "fare","age", edgecolor="b").add_legend()
```

```python
# creating the stripplot
# strip plot is the combination of bar plot and scatter plot
sns.stripplot(data="kashti", x="fare" , y="age")
```

```python
# creating the heatmap plot
corr = kashti.corr()
plt.figure(figsize=(8,8))
sns.heatmap(corr)
```

```python
#Creating the voilin plot
sns.violinplot(data=kashti, x="fare", y="age")
```
