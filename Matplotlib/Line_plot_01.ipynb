#### Line plot

```python
#single line
y = np.array([0,1,2,3,4,5])
plt.plot(y,np.sin(y))
plt.show()
```

```python
#Multiple line
y = np.array([0,1,2,3,4,5])
plt.plot(y,np.sin(y))
plt.plot(y,np.cos(y))
plt.show()
```

```python
#Adjusting the plot: Line and colors
y = np.array([0,1,2,3,4,5])
plt.plot(y,np.sin(y), color="red")
plt.plot(y,np.cos(y), color="black")
plt.show()
```

```python
#Adjusting the lines
y = np.array([0,1,2,3,4,5])
plt.plot(y,np.sin(y), linestyle="dashed")
plt.plot(y,np.cos(y), linestyle="dotted")
plt.show()
```

```python
# Adjusting the plots: Axes Limits
y = np.array([0,1,2,3,4,5])
plt.plot(y,np.sin(y))
plt.plot(y,np.cos(y))
plt.xlim(0,3)
plt.ylim(-1.00,0.00)
plt.show()
```

```python
#Labeling
plt.plot(y,np.sin(y), '-g',label='sin(y)')
plt.plot(y,np.cos(y), ':b',label='cos(y)')
plt.axis()
plt.legend()
```
